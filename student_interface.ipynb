```python

import numpy as np
import random
import math

row_count = 6
col_count = 7

ai_turn = 0
player_turn = 1

player_disk = 1
opp_disk = 2

#will create the board 
def create_board(self,):
    board = np.zeros((6, 7))
    return board 

def is_valid_location(board):
    board[row_count - 1][col] == 0

#finds all possible locations to drop a disk
def get_valid_location(board):
    valid_locations = []
    for col in range(col_count):
        if is_valid_location(board, col):
            valid_locations.append(col)
    return valid_locations

#gets next available row as it exits once it returns i eg 0-6 so starts from lowest number 
def get_next_row(board, col):
    for i in range[6]:
        if board[i][col] == 0:
            return i 

#used to define how to drop a disk
def drop_piece(board, row, count, disk): 
    board[row][count] == disk

#provides score based on number of counters in row 
def window_frame(window, disk): 
    score = 0
    opp_player = player_disk
    if disk == player_disk:
        opp_piece = ai_disk
        
    if window.count(disk) == 4:
        score += 100
    elif window.count(disk) == 3 and window.count(0) == 1:
        score += 5
    elif window.count(disk) == 2 and window.count(0) == 2:
        score += 2 
    #favours blocking an opponent, but less than choosing a winning move 
    if window.count(opp_piece) == 3 and window.count(disk) == 1:
        score -= 4


def score_position(board, disk):
    score = 0

    #scores centre column
    centre_array = [int(i) for i in list(board[:, col_count // 2])]
    centre_count = centre_array.count(disk)
    score += centre_count * 3

    #scores horizontal positions
    for r in range(row_count):
        row_array = [int(i) for i in list(board[r, :])]
        for c in range(col_count - 3):
            # forms horizontal window of 4
            window = row_array[c:c + 4]
            score += window_frame(window, disk)

    #scores vertical positions
    for c in range(col_count):
        col_array = [int(i) for i in list(board[:, c])]
        for r in range(row_count - 3):
            # forms vertical window of 4
            window = col_array[r:r + 4]
            score += window_frame(window, disk)

    #scores positive diagonals
    for r in range(row_count - 3):
        for c in range(col_count - 3):
            # forms positive diagonal window of 4
            window = [board[r + i][c + i] for i in range(4)]
            score += window_frame(window, disk)

    #scores negative diagonals
    for r in range(row_count - 3):
        for c in range(col_count - 3):
            # forms negative diagonal window of 4
            window = [board[r + 3 - i][c + i] for i in range(4)]
            score += window_frame(window, disk)

    return score

def winning_move(board, disk):
#look for horizontal locations for win
    for c in range(col_count - 3):
        for r in range(row_count - 3):
            if board[r][c] == disk and board[r][c + 1] == disk and board[r][c + 2] == disk and board[r][c + 3] == disk:
                return True

    #look for vertical locations for win
    for c in range(col_count):
        for r in range(row_count - 3):
            if board[r][c] == disk and board[r + 1][c] == disk and board[r + 2][c] == disk and board[r + 3][c] == disk:
                return True

    #look for positive diagonal locations for win
    for c in range(col_count - 3):
        for r in range(row_count - 3):
            if board[r][c] == disk and board[r + 1][c + 1] == disk and board[r + 2][c + 2] == disk and board[r + 3][c + 3] == disk:
                return True

    #look for negative diagonal locations for win
    for c in range(col_count - 3):
        for r in range(3, row_count -3):
            if board[r][c] == disk and board[r - 1][c + 1] == disk and board[r - 2][c + 2] == disk and board[r - 3][c + 3] == disk:
                return True
def winning_move(board, disk):
#look for horizontal locations for win
    for c in range(col_count - 3):
        for r in range(row_count - 3):
            if board[r][c] == disk and board[r][c + 1] == disk and board[r][c + 2] == disk and board[r][c + 3] == disk:
                return True

    #look for vertical locations for win
    for c in range(col_count):
        for r in range(row_count - 3):
            if board[r][c] == disk and board[r + 1][c] == disk and board[r + 2][c] == disk and board[r + 3][c] == disk:
                return True

    #look for positive diagonal locations for win
    for c in range(col_count - 3):
        for r in range(row_count - 3):
            if board[r][c] == disk and board[r + 1][c + 1] == disk and board[r + 2][c + 2] == disk and board[r + 3][c + 3] == disk:
                return True

    #look for negative diagonal locations for win
    for c in range(col_count - 3):
        for r in range(3, row_count -3):
            if board[r][c] == disk and board[r - 1][c + 1] == disk and board[r - 2][c + 2] == disk and board[r - 3][c + 3] == disk:
                return True
            
def is_terminal_node(board):
    return winning_move(board, player_disk) or winning_move(board, ai_disk) or len(get_valid_location(board)) == 0

def minimax(board, depth, alpha, beta, maximisingPlayer):
    valid_locations = get_valid_location(board)

    is_terminal = is_terminal_node(board)
    if depth == 0 or is_terminal:
        if is_terminal:
            #give high points to bot for winning 
            if winning_move(board, ai_disk):
                return (None, 9999999)
            #give low points to human for winning 
            elif winning_move(board, player_disk):
                return (None, -9999999)
            else:  #no other moves possible
                return (None, 0)
        # return bot's score
        else:
            return (None, score_position(board, ai_disk))

    if maximisingPlayer:
        value = -9999999
        # randomise column to start
        column = random.choice(valid_locations)
        for col in valid_locations:
            row = get_next_row(board, col)
            b_copy = board.copy()
            # drop piece in temporary board and record score
            drop_piece(b_copy, row, col, ai_disk)
            new_score = minimax(b_copy, depth - 1, alpha, beta, False)[1]
            if new_score > value:
                value = new_score
                # make 'column' the best scoring column we can get
                column = col
            alpha = max(alpha, value)
            if alpha >= beta:
                break
        return column, value

    else:  # Minimising player
        value = 9999999
        # randomise column to start
        column = random.choice(valid_locations)
        for col in valid_locations:
            row = get_next_row(board, col)
            b_copy = board.copy()
            # drop piece in temporary board and record score
            drop_piece(b_copy, row, col, player_disk)
            new_score = minimax(b_copy, depth - 1, alpha, beta, True)[1]
            if new_score < value:
                value = new_score
                # make 'column' best scoring column we can get
                column = col
            beta = min(beta, value)
            if alpha >= beta:
                break
        return column, value
    
board = create_board(([[int]*6]*7))
turn = random.randint(0,6)

game_not_over = False

while game_not_over:
    if turn == ai_turn:
        col, minimax_score = (board, 5, -math.inf, math.inf, True)
```
